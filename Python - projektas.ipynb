```python
from csv import DictReader

info = []

with open('Happiness_index.csv', encoding='utf8') as failas:
    readeris = DictReader(failas, delimiter=';')

    for eilute in readeris:
        eilute['Year'] = int(eilute['Year'])
        eilute['Happiness index'] = float(eilute['Happiness index'])
        eilute['Pollution'] = float(eilute['Pollution'])
        info.append(eilute)

print('Visi turimi duomenys: ')
for i in info:
    print(i)
```

```python
lietuva = [l for l in info if l['Country'] == 'Lithuania']
suomija = [l for l in info if l['Country'] == 'Finland']
danija = [l for l in info if l['Country'] == 'Denmark']
norvegija = [l for l in info if l['Country'] == 'Norway']
latvija = [l for l in info if l['Country'] == 'Latvia']
estija = [l for l in info if l['Country'] == 'Estonia']

print('Latvija: ')
for inf in latvija:  
    print(inf)
```

```python
vidurkis_laimes = round( sum(i['Happiness index'] for i in info)/len(info),2)
print('Stebimo laikotarpio laimes indekso vidurkis: ', vidurkis_laimes)

vidurkis_uzterstumo = round( sum(i['Pollution'] for i in info)/len(info),2)
print('Stebimo laikotarpio uzterstumo vidurkis: ', vidurkis_uzterstumo)
print()
min_laimes = min(i['Happiness index'] for i in info)
max_laimes = max(i['Happiness index'] for i in info)
print('Maziausias laimes indeksas: ', min_laimes, '\nDidziausias laimes indeksas: ',max_laimes)
print()
min_uzters = min(i['Pollution'] for i in info)
max_uzters = max(i['Pollution'] for i in info)
print('Maziausias uzterstumas: ', min_uzters, '\nDidziausias uzterstumas: ',max_uzters)
```

```python
nelaimingos = []
laimingos = []
for i in info:
    if i['Happiness index'] <= vidurkis_laimes:
        nelaimingos.append(i['Country'])
    else:
        laimingos.append(i['Country'])

nelaimingos = set(nelaimingos)
laimingos = set(laimingos)

print('Salys, kuriose laimės indeksas nesiekia vidurkio:')
print(nelaimingos)
print('Salys, kuriose laimės indeksas virsina vidurki:')
print(laimingos)
```

```python
neuzterstos = []
uzterstos = []
for i in info:
    if i['Pollution'] <= vidurkis_uzterstumo:
        neuzterstos.append(i['Country'])
    else:
        uzterstos.append(i['Country'])

neuzterstos = set(neuzterstos)
uzterstos = set(uzterstos)

print('Salys, kuriose uzterstumas, bent vienais is stebimu metu, nesiekia vidurkio:')
print(neuzterstos)
print('Salys, kuriose uzterstumas, bent vienais is stebimu metu, virsina vidurki:')
print(uzterstos)
```
